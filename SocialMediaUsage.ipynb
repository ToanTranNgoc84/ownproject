# 3 Data Cleaning

## #3.1 Format Column Names

```python
#Remove spaces and lower letters
df1.columns = df1.columns.str.lower().str.replace(" ", "_")
df2.columns = df2.columns.str.lower().str.replace(" ", "_")

#Rename column names
df1.rename(columns={"user_id": "id"}, inplace=True)
df2.rename(columns={"user_id": "id"}, inplace=True)

print()
display(df1)
display(df2)
```

## 3.2 Format Columns

### 3.2.1 Dataset 1: Replace "Male" in column "age"

```python
#Replace "Male" wir NaN in df1
df1["age"].replace({
    "Male": np.nan,   
     # Replace '.' with NaN
}, inplace=True)
df1["age"].fillna("Unknown", inplace=True)
```

### 3.2.2 Dataset 1: Replace "Female", "Non-binary", "Male", "27", "Marie" in column "gender" for "F", "M", "N"

```python
df1["gender"].replace({
    "Female": "F",
    "Non-binary": "D",
    "Male": "M",
    "27": np.nan,
    "Marie": np.nan
}, inplace=True
)
```

### 3.2.3 Dataset 2: Replace "Male", "Female", "Non-binary" in "age"

```python
#Replace "Male", "Female", "Non-binary" in "age"
df2["age"].replace({
    "Male": np.nan,
    "Non-binary": np.nan,
    "Female": np.nan,
    " işte mevcut veri kümesini 1000 satıra tamamlıyorum:": np.nan,
}, inplace=True
)
```

### 3.2.4 Dataset 2: Replace "Female", "Non-binary", "Male", and all numbers for "F", "N", "M", NaN

```python
df2["gender"].replace({
    "Female": "F",
    "Non-binary": "D",
    "Male": "M",
    "27": np.nan,
    "24": np.nan,
    "29": np.nan,
    "33": np.nan,
    "31": np.nan,
    "25": np.nan,
    "28": np.nan,
    "23": np.nan,
    "34": np.nan,
    "35": np.nan,
    "21": np.nan,
    "32": np.nan,
    "22": np.nan,
    "30": np.nan,
    "23": np.nan,
    "26": np.nan
}, inplace=True
)
```
